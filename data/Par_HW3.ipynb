```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/726 Parallel/problem_three.csv')
data_3
```

```python
fig = plt.figure(figsize=(10, 10))              
x_label = data_3['bc']
y_label = data_3['speedup']
plt.plot(x_label, y_label, color = 'r',linewidth=1.0,linestyle=':')  
plt.title("Speedup & bc")                 
plt.ylabel("speedup")            
plt.xlabel("bc")            #
plt.show()   
```

```python
data_5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/726 Parallel/problem_five.csv')
data_5
```

```python
fig = plt.figure(figsize=(20, 20))  
##########################
x_label = data_5['bc']  
y_label = data_5['speedup']     
plt.subplot(2,2,1)         
plt.plot(x_label, y_label, color = 'r',linewidth=1.0,linestyle=':')  
plt.title("Speedup & bc")                 
plt.ylabel("speedup")            
plt.xlabel("bc")                         
#########################
x_label = data_5['bc']  
y_label = data_5['efficiency']     
plt.subplot(2,2,2)         
plt.plot(x_label, y_label, color = 'b',linewidth=1.0,linestyle=':')  
plt.title("Efficiency & bc")                 
plt.ylabel("efficiency'")            
plt.xlabel("bc")                 
##########################
x_label = data_5['num_threads']  
y_label = data_5['speedup']     
plt.subplot(2,2,3)         
plt.plot(x_label, y_label, color = 'r',linewidth=1.0,linestyle=':')
plt.xticks(np.arange(1, 45, 1), rotation=60)
plt.title("Speedup & num_threads")                 
plt.ylabel("speedup")            
plt.xlabel("num_threads")                         
#########################
x_label = data_5['num_threads']  
y_label = data_5['efficiency']     
plt.subplot(2,2,4)         
plt.plot(x_label, y_label, color = 'b',linewidth=1.0,linestyle=':')
plt.xticks(np.arange(1, 45, 1), rotation=60) 
plt.title("Efficiency & num_threads")                 
plt.ylabel("efficiency")            
plt.xlabel("num_threads")            
plt.show()  
```
